```python
# Create training and testing arrays
training = np.array(data1,dtype='float32')
testing = np.array(data2,dtype ='float32')
```

```python
# Prepare the training and testing dataset 
X_train = training[:,1:]/255
y_train = training[:,0]

X_test = testing[:,1:]/255
y_test = testing[:,0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, test_size = 0.3, random_state = 12345)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
print('Training Data Shape   : ', X_train.shape)
print('Training Labels Shape : ', y_train.shape)
print('Testing Data Shape    : ', X_test.shape)
print('Testing Labels Shape  : ', y_test.shape)
```

```python
# Training The Model
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 8)
classifier.fit(X_train, y_train)
```

```python
# Evaluate the model
pred_train = classifier.predict(X_train)
pred_test  = classifier.predict(X_test)

from sklearn.metrics import accuracy_score
print('Training Set Accuracy : ', accuracy_score(y_train, pred_train))
print('Testing Set Accuracy  : ', accuracy_score(y_test, pred_test))
```

```python
classifier.predict(X_test[:5])
```

```python
# Training A decision tree model
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(max_depth = 10)
classifier.fit(X_train, y_train)
```

```python
# Evaluate the model
pred_train = classifier.predict(X_train)
pred_test  = classifier.predict(X_test)

from sklearn.metrics import accuracy_score
print('Training Set Accuracy : ', accuracy_score(y_train, pred_train))
print('Testing Set Accuracy  : ', accuracy_score(y_test, pred_test))
```

###  Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, max_depth = 10, criterion = 'squared_error')
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

pred_train = regressor.predict(X_train)
pred_test  = regressor.predict(X_test)

print('Training Set Score : ', r2_score(y_train, pred_train))
print('Testing Set Score  : ', r2_score(y_test, pred_test))
```

### XGB Regressor

```python
from xgboost import XGBRegressor
regressor = XGBRegressor(n_estimators = 50, max_depth = 3)
regressor.fit(X_train, y_train)
```

```python
pred_train = regressor.predict(X_train)
pred_test  = regressor.predict(X_test)

print('Training Set Score : ', r2_score(y_train, pred_train))
print('Testing Set Score  : ', r2_score(y_test, pred_test))
```

### LGBM BOOSTER

```python
from lightgbm import LGBMRegressor
regressor = LGBMRegressor(n_estimators = 50)
regressor.fit(X_train, y_train)
```

```python
pred_train = regressor.predict(X_train)
pred_test  = regressor.predict(X_test)

print('Training Set Score : ', r2_score(y_train, pred_train))
print('Testing Set Score  : ', r2_score(y_test, pred_test))
```

### ADABOOST Regressor

```python
from sklearn.tree import DecisionTreeRegressor
estimator = DecisionTreeRegressor(max_depth = 5, criterion = 'squared_error')

from sklearn.ensemble import AdaBoostRegressor
regressor = AdaBoostRegressor(estimator, n_estimators = 50)
regressor.fit(X_train, y_train)
```

```python
pred_train = regressor.predict(X_train)
pred_test  = regressor.predict(X_test)

print('Training Set Score : ', r2_score(y_train, pred_train))
print('Testing Set Score  : ', r2_score(y_test, pred_test))
```
