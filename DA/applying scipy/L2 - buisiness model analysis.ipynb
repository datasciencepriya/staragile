##### 1. Douniversities with a high number of applications are the preferred ones by students? in other words, could the number of applications tell us that a university isone of the most preferred by students?

```python
number_of_applications =  university.groupby('Name')['Applicants total'].sum()
number_of_applications.sort_values(ascending=False)
```

#### another way:

```python
number_of_applications = university[['Name','Applicants total']].sort_values("Applicants total",ascending = False).head(10)
number_of_applications
```

```python
plt.figure(figsize=(20,16))
sns.barplot(x='Applicants total',y='Name',data =number_of_applications)
plt.title('Top American Universities with the Most Applications')
plt.xlabel('Number of applications')
```

### Top american universities more than 20000 application

```python

filtered_data = number_of_applications[number_of_applications['Applicants total'] > 20000]

plt.figure(figsize=(20, 16))

sns.barplot(x='Applicants total', y='Name', data=filtered_data)
plt.title('Top American Universities with More Than 20000 Applications')
plt.xlabel('Number of Applications')
plt.ylabel('University Name')

plt.show()

```

 the universities with the highest number of applications are the preferred ones by students. However, one student can send several applications at the same time, but only one university could be the preferred one. let's find some relationships with the number of admissions and enrollments.

```python
plt.figure(figsize=(16,6))

plt.subplot(1,3,1)
sns.histplot(university['Applicants total'],bins=50)
plt.title('Histogram of Number of Applications. ')
plt.xlabel('Number of Applications')
plt.axis([0,30000,0,350])
plt.xticks(rotation=45)


plt.subplot(1,3,2)
sns.histplot(university['Admissions total'],bins=50)
plt.title('Histogram of Number of Admissions.')
plt.axis([0,10000,0,350])
plt.xlabel('Number of Admissions')
plt.xticks(rotation=45)


plt.subplot(1,3,3)
sns.histplot(university['Enrolled total'],bins=50)
plt.title('Histogram of Number of Enrollments.')
plt.axis([0,5000,0,350])
plt.xlabel('Number of Enrollments')

plt.xticks(rotation=45)
```

As we can see, the majority of universities received less than 5000 applications. However, some universities received a much higher number of applications, as many as 30000 applications in the same year.

Talking about enrollments, we'll find lower numbers when compared to the number of applications and admissions. That happens because students usually apply to several universities at the same time, hoping that the favorite one admits them; however, when that is not the case, they still have other options.

### To know the admission rate 

```python

plt.figure(figsize=(16, 6))

# Subplot 1
plt.subplot(1, 2, 1)
plt.title('APPLICATIONS VS ADMISSIONS')
sns.scatterplot(x=university['Applicants total'], y=university['Admissions total'], hue=university['Control of institution'])
plt.xlabel('Number of Applications')
plt.ylabel('Number of Admissions')
plt.grid()

# Subplot 2
plt.subplot(1, 2, 2)
plt.title('ADMISSIONS VS ENROLLMENTS')
sns.scatterplot(x=university['Admissions total'], y=university['Enrolled total'], hue=university['Control of institution'])
plt.xlabel('Number of Admissions')
plt.ylabel('Number of Enrollments')
plt.grid()

plt.tight_layout(pad=2)
plt.show()
```

### correlation exists between high number of applicatoions and high admissions

```python
university['acceptance_rate'] = (university['Admissions total']/university['Applicants total'] * 100)
university['enrollment_rate'] = (university['Enrolled total']/university['Admissions total'] * 100)
```

```python
plt.figure(figsize = (12,5))
sns.scatterplot(x = 'Applicants total', y = 'enrollment_rate', data= university)
plt.xlabel('Number of Applications')
plt.ylabel('Enrollemnt rate %')
plt.grid()
```

```python
plt.figure(figsize = (16,6))
sns.scatterplot(x = university.acceptance_rate,y = university.enrollment_rate, hue = university['Control of institution'])
plt.xlabel('Acceptance Rate')
plt.ylabel('Enrollment Rate')
plt.grid()
```

```python
high_acceptance = university[university.acceptance_rate.notnull()][['Name','acceptance_rate','enrollment_rate']].sort_values('acceptance_rate',ascending = False).head(10)
```

```python
low_acceptance = university[university.acceptance_rate.notnull()][['Name','acceptance_rate','enrollment_rate']].sort_values('acceptance_rate', ascending= True).head(10)
```

```python
plt.figure(figsize = (16,4))

plt.subplot(1,2,1)
ind = np.arange(len(high_acceptance))
width = 0.30

plt.bar(ind, high_acceptance.acceptance_rate, width, label = 'Acceptance Rate')
plt.bar(ind+ width, high_acceptance.enrollment_rate, width, label = 'Enrollment Rate')
plt.title('Top 10 Univeristies with Highest Acceptance Rate')
plt.ylabel('Rates %')
plt.xticks(ind + width, high_acceptance.Name.values, rotation = 90)
plt.legend(loc = 'best')

plt.subplot(1,2,2)
ind = np.arange(len(low_acceptance))
width = 0.30

plt.bar(ind, low_acceptance.acceptance_rate, width, label = 'Acceptance Rate')
plt.bar(ind+ width, low_acceptance.enrollment_rate, width, label = 'Enrollment Rate')
plt.title('Top 10 Univeristies with Lowest Acceptance Rate')
plt.ylabel('Rates %')
plt.xticks(ind + width, low_acceptance.Name.values, rotation = 90)
plt.legend(loc = 'best')
```

The data does not reveal a distinct relationship between high admission rates and enrollment rates. This suggests that student preferences for universities are not primarily determined by the ease of admission

##### 8. Dostudents prefer a university based on the possibility of a higher, additional academic degree in the same university?

```python
df = pd.read_csv('universities_data.csv')

df.head(5)
```

```python
higher_degree = ["Offers Master's degree",
                "Offers Doctor's degree - research/scholarship",
                "Offers Doctor's degree - professional practice",
                "Offers Doctor's degree - other"]

for degree in higher_degree:
    sns.boxplot(data=df, x=degree, y='Admissions yield - total')
    plt.show()
```

```python
degree_offers = ['Offers Less than one year certificate',
                           'Offers One but less than two years certificate', 
                           "Offers Associate's degree",
                           "Offers Two but less than 4 years certificate",
                           'Offers Postbaccalaureate certificate',
                           "Offers Master's degree",
                           "Offers Post-master's certificate",
                           "Offers Doctor's degree - research/scholarship",
                           "Offers Doctor's degree - professional practice",
                           "Offers Doctor's degree - other",
                            'Offers Other degree']

df_n = df[degree_offers + ['Admissions yield - total']]


df_n.head()
```

```python
other_degrees = list(set(degree_offers) - set(higher_degree))
other_degrees
```

```python
for degree in other_degrees:
    sns.boxplot(data=df_n, x=degree, y='Admissions yield - total')
    plt.show()
```

students prefer a university based on the possibility of a higher, additional
academic degree in the same university
